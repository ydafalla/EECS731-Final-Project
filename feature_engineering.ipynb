```python
def uppercase_finder(string):
    uppercase_index = 0
    for i in string:
          if (i.isupper()) == True:
              uppercase_index = 1
              break
    return uppercase_index
```

```python
def uppercase_counter(string):
    uppercase_count= 0
    for i in string:
          if(i.isupper()):
                uppercase_count = uppercase_count + 1
    return uppercase_count
    
```

```python
def lowercase_count(string):
    count=0
    for i in string:
        if(i.islower()):
            count+=1
    return count
```

```python
def numeric_finder(string):
    numeric=0
    for i in string:
        if(i.isnumeric()):
            numeric=1
            break
    return numeric
            
```

```python
def numeric_count(string):
    count=0
    for i in string:
        if(i.isnumeric()):
            count+=1
    return count
```

```python
import re
def specialchar_finder(string):
    index=0
    regex=re.compile('[@_!#$%^&*( )<>?/\|}{~:]')
    if(regex.search(string)==None):
        index=0
    else:
        index=1
    return index
```

```python
def special_char_count(string):
    count=0
    for i in string:
        if(not i.isalnum()):
            count+=1
    return count
```

```python
def alpha_count(string):
    count=0
    for i in string:
        if(i.isalpha()):
            count+=1
    return count
```

```python
def origin(string):
    yahoo=pass_list1[0].values.tolist()
    hotmail=pass_list2[0].values.tolist()
    netflix=pass_list2[0].values.tolist()
    if(string in yahoo):
        return 'Yahoo'
    elif(string in hotmail):
        return 'Hotmail'
    else:
        return 'Netflix'
```

```python
new_df['Has_Upper']=new_df['Passwords'].apply(uppercase_finder)
new_df['Has_Numeric']=new_df['Passwords'].apply(numeric_finder)
new_df['Has_SpecialChar']=new_df['Passwords'].apply(specialchar_finder)
new_df['UpperCase_Count']=new_df['Passwords'].apply(uppercase_counter)
new_df['LowerCase_Count']=new_df['Passwords'].apply(lowercase_count)
new_df['Numeric_Count']=new_df['Passwords'].apply(numeric_count)
new_df['SpecialChar_Count']=new_df['Passwords'].apply(special_char_count)
new_df['Alphabet_Count']=new_df['Passwords'].apply(alpha_count)
```

```python
def f(string):
    val=0
    if(len(string)<=6 or string.isnumeric()):
        val=1
    else:
        if(string.isalpha()):
            val=2
        else:
            val=3
    return val
```

```python
def has_word(string):
    new_words=words[0].values.tolist()
    for word in new_words:
        if(word in string):
            return 1
    return 0
```

```python
def has_name(string):
    new_names=names[0].values.tolist()
    for name in new_names:
        if(name in string):
            return 1
    return 0
```

```python
def has_place(string):
    new_places=places[0].values.tolist()
    for place in new_places:
        if(place in string):
            return 1
    return 0
```
